- FaceNet, VGG-Face, ArcFace and Dlib are overperforming ones based on experiments

```python
from deepface import DeepFace
backends = ['mtcnn','retinaface']
models = ['Facenet512', 'Facenet', 'VGG-Face', 'Dlib']
detector = backends[0]
model = models[2]
align = True
image = "../dataset/raw/DSC00764.JPG"
embedding_objs = DeepFace.represent(
  img_path = image, detector_backend = detector, align = align, model_name = models[0]
)
print(embedding_objs)
```
